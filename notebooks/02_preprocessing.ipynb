# Data Preprocessing

This notebook focuses on the preprocessing steps required to prepare financial text data for sentiment analysis. Before any machine learning model can learn meaningful patterns, the raw text must be cleaned, normalized, and transformed into a numerical representation. The goal of this preprocessing pipeline is to create a consistent and reproducible workflow that converts unstructured financial phrases into features suitable for classical machine learning models.

The steps implemented here include lowercasing, tokenization, removal of unwanted characters, lemmatization, and stopword filtering. These operations help reduce noise while preserving the linguistic signals that matter for sentiment classification.

```python
import sys
from pathlib import Path

project_root = Path().resolve()
while not (project_root / "src").exists():
    project_root = project_root.parent

sys.path.append(str(project_root))
```

```python
from src.utils.utils import load_config
from src.data.make_dataset import load_txt
from src.data.clean_data import make_cleaning_pipeline
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer


nltk.download('stopwords')
stop_words = stopwords.words('english')
nltk.download('wordnet')
lemmatizer= WordNetLemmatizer()
```

# Cleaning the text data

```python
config = load_config("configs/config.yaml")

df = load_txt(config['data']['raw_path'], names=['text', 'label'])

clean_pipe = make_cleaning_pipeline(stop_words=stop_words, lemmatizer=lemmatizer)

df["clean_text"] = clean_pipe.transform(df["text"])

df.head()
```

```python
df.to_csv(config['data']['preprocessed_path'], index=False, na_rep="")
```
